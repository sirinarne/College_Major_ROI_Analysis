# College Major ROI Analysis

## Project Overview

Choosing a college major is a crucial part of teenagers' lives. This decision is especially important financially speaking. This project analyzes labor market outcomes across college majors using salary and employment data. The goal is to identify which majors provide the strongest return on investment (ROI) by examining earnings, unemployment rates, and employment outcomes.

### Research Questions

1. Which majors have the highest median salaries?
2. Which majors have the lowest unemployment rates?
3. How does salary vary across major categories?
4. Is there a relationship between salary and unemployment?
5. Which majors provide the strongest ROI when both salary and employment outcomes are considered?

## ROI Metric Development 

- This ROI metric rewards higher salaries and penalizes higher unemployment 
- The formula is: 

$$ 
ROI\_Score = \frac {Median\ Salary}{1+Unemployment\ Rate}
$$

Where: 
- Median Salary = median earnings for the major
- Unemployment Rate = proportion of graduates unemployed for the major 


```python
df["ROI_Score"]  = (df["Median"]/(1+df['Unemployment_rate']))
```

```python
df[["Major", "Major_category","Median","Unemployment_rate","ROI_Score"]].sort_values("ROI_Score",ascending = False).head(20)
```

- Petroleum engineering has the highest median salary + highest ROI
- Metallurgical Engineering jumped above Mining (lower unemployment)
- Nuclear engineering fell - unemployment is high (17.7%)
- Engineering Mechanics Physics and Science jumped because unemployment is only 0.6%
- Shows that salary alone does not fully capture career outcomes; employment stability should also be considered. 

## ROI Analysis

```python
category_roi = (df.groupby("Major_category")["ROI_Score"].mean().sort_values(ascending = False))
category_roi
```

- Engineering and Business remained at number 1 and 2 spots, respectively 
- Physical Sciences jumped above law + computers
- Computers/mathematics and law both dropped 

```python
top_roi = df.nlargest(15, "ROI_Score")
plt.figure(figsize=(10,6))
plt.barh(top_roi["Major"],top_roi["ROI_Score"])
plt.title("Top 15 majors by ROI Score")
plt.xlabel("ROI_Score")
plt.tight_layout()
plt.show()
```

## Findings

1. Engineering majors consistently ranked highest in both salary and ROI analyses.

2. Business and Computer/Mathematics majors also demonstrated strong earnings outcomes.

3. Salary distributions were right-skewed, with a small number of majors earning substantially more than the average.

4. Salary and unemployment exhibited only a weak negative relationship.

5. Several majors achieved strong ROI rankings due to a combination of high salaries and low unemployment rates.

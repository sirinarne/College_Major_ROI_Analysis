```python
#The following majors have the highest median salaries - many of them are engineering
df.sort_values("Median", ascending=False)[["Major", "Median"]].head(10)
```

```python
#Salary Gap: 75th percentile - 25th percentile
df["salary_gap"] = df["P75th"] - df["P25th"]
df.sort_values("salary_gap", ascending = False)[['Major','salary_gap']].head(10)

#These represents the majors with the widest income spread
#High variability in earnings
```

```python
#Salary Gap: 75th percentile - 25th percentile
df["salary_gap"] = df["P75th"] - df["P25th"]
df.sort_values("salary_gap")[['Major','salary_gap']].head(10)

#These represents the majors with the lowest income spread
#Typically more standardized career paths, more consistent pay across graduates
```

```python
# Represents the majors with the highest unemployment rates
df.sort_values("Unemployment_rate", ascending = False)[["Major", "Unemployment_rate"]].head(10)
```

```python
#Represents majors with the lowest unemployment rates
df.sort_values("Unemployment_rate")[["Major", "Unemployment_rate"]].head(10)
```

```python
df["Major_category"].value_counts()
# Top three categories with the most majors: Engineering, education, humanities/liberal arts
# Categories with least amount of majors: Law, communications, interdisciplinary
```

```python
df.groupby("Major_category")["Median"].mean().sort_values(ascending=False)
#Average salary by category - Engineeing, business, and computers/math pay the most 
#Psych, humanities, education pay the least
```

```python
df["Median"].hist()
plt.title("Distribution of Median Salaries")
plt.show()
```

- right skewed: most majors clustered in the middle/lower range
- few outliers - some majors with very high salaries 

```python
#Engineering typically pays high -- proof of outliers (represents outliers on hist. above)
Q1 = df["Median"].quantile(0.25)
Q3 = df["Median"].quantile (0.75)
IQR = Q3-Q1

outliers = df[(df["Median"]<Q1-1.5*IQR) | (df["Median"] > Q3 + 1.5*IQR)]
outliers[["Major", "Median"]]
```

```python
plt.scatter(df['Unemployment_rate'],df["Median"])
plt.xlabel("Unemployment Rate")
plt.ylabel("Median Salary")
plt.title("Salary vs Unemployment")
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
df["Median"].corr(df["Unemployment_rate"])

# weak negative correlation
```

- The correlation between median salary and unemployment rate is -0.12, which is weak. This suggests that there is a very slight tendency for higher-paying majors to have lower unemployment rates. The relationship is not strong enough to conclude that salary is a reliable predictor of unemployment.

-Example: Nuclear engineering has one of the highest median salaries but is also in the top 10 for highest unemployment rates.

```python
df.groupby("Major_category")["Median"].mean().sort_values().plot(kind="barh")
plt.title("Average Median Salary by Major Category")
plt.show()
```

```python
df.groupby("Major_category")["Unemployment_rate"].mean().sort_values().plot(kind="barh")
plt.title("Average Unemployment Rate by Major Category")
plt.show()
```

KEY FINDINGS:
1. Engineering, business, and computer/math majors generally have the highest median salaries
2. Education, psychology, and humanities majors tend to have lower salaries
3. Salaries are right - skewed; there are a small number of majors earning substantially more than the rest
4. Unemployment rates vary across majors, but the relationship between salary and unemployment is relatively weak.
5. Some engineering/STEM majors have high median salaries along with low unemployment rates, making them some of the strongest - performing majors in the dataset.

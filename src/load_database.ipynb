```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("../college_roi.db")
```

```python
dashboard_df = pd.read_csv(
    "../data/processed/dashboard_data.csv"
)
```

```python
dashboard_df.to_sql(
    "majors",
    conn,
    if_exists="replace",
    index=False
)
```

```python
conn.close()

print("Database created successfully!")
```
